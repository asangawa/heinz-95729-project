```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
cervical_cancer = pd.read_csv('df_risk_factors_cervical_cancer.csv')
cervical_cancer
```

```python
cervical_cancer.columns[1:33]
```

```python
biopsy_positive = cervical_cancer[cervical_cancer['Biopsy'] == 1]
biopsy_negative = cervical_cancer[cervical_cancer['Biopsy'] == 0]
colors=['b', 'r']
for feature in cervical_cancer.columns[1:33]:
    print('Distribution of {}_biopsy positive'.format(feature))
    plt.hist(biopsy_positive[feature], bins = 10)
    plt.show()
    print('Distribution of {}_biopsy negative'.format(feature))
    plt.hist(biopsy_negative[feature])
    plt.show()
#     plt.legend((blue, red), ('biopsy_positive', 'biopsy_negative'))
    
```
